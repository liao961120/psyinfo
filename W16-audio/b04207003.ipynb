```python
import matplotlib.pyplot as plt
import numpy
urllib.request.urlretrieve('https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml','haarcascade_frontalface_default.xml')
urllib.request.urlretrieve('https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_eye.xml','haarcascade_eye.xml')

def detect_face(img):
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
    faces = face_cascade.detectMultiScale(img)
    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = img[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));
```

```python
# Detect face in cartoon
detect_face(cartoon)
```

```python
### Detect face in cloud
detect_face(cloud)
```

Please use one audio/speech segementation method of your choice (including your own codes) to calculate:

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
%matplotlib inline
y, sr = librosa.load('happy/53_F_H_6.wav', sr=None, mono=True) # sr=none: ‘None’ uses the native sampling rate, mono=True 單聲道
#pl = librosa.display.waveplot(y, sr);
```

```python
def count_seg(file, threshold=25):
    # Load audio file
    y, sr = librosa.load(file, sr=None, mono=True) # sr=none: ‘None’ uses the native sampling rate, mono=True 單聲道
    # get segment ranges
    split_y=librosa.effects.split(y, top_db=threshold)
    # num of segments
    nseg = split_y.shape[0]
    # total speech duration
    total_duration = .0
    for i in range(nseg):
        total_duration += librosa.get_duration(y[split_y[i][0]:split_y[i][1]])

    return nseg, total_duration
```

```python
import glob
file_list = glob.glob("happy/*.wav") + glob.glob('anger/*.wav')
for fp in file_list:
    nseg, duration = count_seg(fp, 20)
    print(" %s\t Number of Segments: %d\t Duration: %2.3f sec" % (fp, nseg, duration))
```
